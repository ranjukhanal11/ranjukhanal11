We took the data source from the WHO website. THe link of dataset is given below. The data was cleaned and it was not required to clean the dataset. Thats why we didn't apply the data cleaning.

```python
covid19=pd.read_csv('https://covid19.who.int/WHO-COVID-19-global-data.csv', parse_dates=['Date_reported'])
```

```python
covid19.head()
```

```python
allcovid19=covid19.groupby(['Date_reported']).sum()
covid19['Total Cases']= covid19[['Cumulative_cases', 'Cumulative_deaths']].sum(axis=1)
ax=allcovid19.plot(figsize=(25,8))
ax.set_xlabel('Date_reported')
ax.set_ylabel('# of Cases in the world')
ax.title.set_text('Covid 19 analysis in the world')
plt.show()
```

In above diagram, i plot a graph showing the number of cases in worldwide.

Next, we are going to find the trends in different countries using group by date.

```python
India_covid19=covid19[covid19['Country']== 'India'].groupby(['Date_reported']).sum()
US_covid19=covid19[covid19['Country']== 'United States of America'].groupby(['Date_reported']).sum()
Brazil_covid19=covid19[covid19['Country']== 'Brazil'].groupby(['Date_reported']).sum()
Russia_covid19=covid19[covid19['Country']== 'Russian Federation'].groupby(['Date_reported']).sum()
SA_covid19=covid19[covid19['Country']== 'South Africa'].groupby(['Date_reported']).sum()
fig=plt.figure(figsize=(20,5))
ax=fig.add_subplot(111)
ax.plot(US_covid19[['Total Cases']],label='US')
ax.plot(India_covid19[['Total Cases']],label='India')
ax.plot(Brazil_covid19[['Total Cases']],label='Brazil')
ax.plot(Russia_covid19[['Total Cases']],label='Russia')
ax.plot(SA_covid19[['Total Cases']],label='South Africa')
ax.set_xlabel('Date_reported')
ax.set_ylabel('# of Total Cases')
ax.title.set_text('Top 5 Covid 19 Cases by Country')
plt.legend(loc='upper left')

plt.show()
```

In our next step, we are going to create the stacled bar graph of new cases. We are going to create this graph coutry wise but in the actual graph it was region wise. so this would make our visualization different from the actual one.

```python
fig=plt.figure(figsize=(20,8))
ax=fig.add_subplot(111)
plt.yscale("log")

ax.bar(US_covid19['Date_reported'],US_covid19['New_cases'], color='g',label='US Daily New Cases')
ax.bar(India_covid19['Date_reported'],India_covid19['New_cases'], color='r',label='India Daily New Cases')
ax.bar(Brazil_covid19['Date_reported'],Brazil_covid19['New_cases'], color='pink',label='Brazil Daily New Cases')
ax.bar(Russia_covid19['Date_reported'],Russia_covid19['New_cases'], color='gray',label='Russia Daily New Cases')

ax.set_xlabel('Date')
ax.set_ylabel('# of Affected people')
ax.title.set_text('Four Countries New Cases Data')
plt.legend(loc='upper left')
plt.show()
```
