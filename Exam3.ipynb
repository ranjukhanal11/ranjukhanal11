```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/worldCupPlayer.csv')
```

**(b) (10 points) Show a table with all the data about your favorite club or nationality. Include some description about
an interesting fact in the data.**

```python
rslt_df = df[df['club'] == 'PSV Eindhoven']
```

```python
rslt_df.head()
```

```python
rslt_df.plot(x ='player', y='Goals', kind = 'bar', title='Goals of players')
```

**(c) (10 points) Create a histogram of the ‘Goals’.**

```python
ax = df.Goals.hist(bins = 20)
ax.set(title = "Histogram of Goals")
ax.set_xlabel('Goals')
ax.set_ylabel('Counts')
plt.show()
```

**(d) (10 points) Create a histogram of the ‘Goals’ and ‘Assists’ on the same axis**

```python
plt.hist(df['Goals'], bins=25, alpha=0.45, color='red')
plt.hist(df['Assists'], bins=25, alpha=0.45, color='blue')
  
plt.title("Goals & Assists Histogram")
  
plt.legend(['Goals', 
            'Assists'])
  
plt.show()
```

**(e) (10 points) Create a barchart for most goals scored by nationality.**

```python
maxGoals = df[df.Goals == df.Goals.max()]
```

```python
maxGoals.head()
```

```python
sns.barplot(x = 'nationality',y = 'Goals',data = maxGoals)
plt.show()
```

**(f) (10 points) Create a boxplot of ‘Yel’ (yellow cards) by nationality**

```python
sns.boxplot(data = df, x = 'nationality', y = 'Yel')
```

**(g) (10 points) Create ‘Goals’ and ‘Assists’ as scatter. Signify your favorite club or nationality in some way on the
scatterplot. You may need to add a new column to the dataset to do so**

```python

import itertools
colors = itertools.cycle(["r", "g"])
ax = sns.scatterplot(data = maxGoals, x = 'Goals', y = 'Assists', hue='nationality', color=next(colors))
ax.set(title = "Goal vs Assists of My Favourite Team")

plt.show()
```

```python
import itertools
colors = itertools.cycle(["r", "g"])
ax = sns.scatterplot(data = df, x = 'Goals', y = 'Assists', hue='nationality', color=next(colors))
ax.set(title = "Goal vs Assists All Teams")

plt.show()
```

**(h) (10 points) Create stacked barplot of goals and assists of French National Team. (they won this tourney!) Be sure
to comment on why it is appropriate to stack goals and assists**

```python
FrenchNational = df[df.nationality == 'France']
```

```python
FrenchNational = FrenchNational[df.nationality == 'France']
```

```python
FrenchNational.head()
```

```python
plt.bar(FrenchNational.player, FrenchNational.Goals, color='r')
plt.bar(FrenchNational.player, FrenchNational.Assists, bottom=FrenchNational.Goals, color='b')
plt.xlabel("French Player")
plt.ylabel("Goals and Assists")
plt.legend(["Goals", "Assists"])
plt.show()
```

Stacked bar graphs are used to highlight how a bigger category is broken into smaller categories and how each element affects the total amount. Stacked bar graphs are divided into two categories: Each value for the segment is placed after the previous one in a simple stacked bar graph.

**(i) (10 points) Create a correlation heatmap of the entire dataset.**


```python
ax = sns.heatmap(df.corr(),vmin = -1)
ax.set(title = 'Heatmap of Players dataset')

plt.show()
```
