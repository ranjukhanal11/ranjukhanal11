##(e) (10 points) Impute ‘total_steps’ by replacing the ‘NaN’s with an appropriate number of steps. Convert to appropriate datatype.

```python
df.total_steps.fillna(value=df.total_steps.mean(), inplace = True)
df.head(10)
```

In abov code segment we replaced NaN values with the mean of that column and in below code section we are going to change its data type into integer. 

```python
df.total_steps = df.total_steps.astype(int)
df.head(5)
```

##(f) (10 points) Convert non-empty ‘avg_pace’ into a float that is still representative of the information contained in the column. Recall that there are 60 seconds in one minute so 3:30 is equivalent to 3.5 minutes.

As we know that data type of avg_pace column is object and now we are going to first convert into datetime. Then to covert minutes into hour , we will divide it with 60.


```python
df.avg_pace = pd.to_datetime(df.avg_pace)
df.avg_pace.dtypes
```

```python
df.avg_pace = df.avg_pace.apply(lambda avg_pace: avg_pace.hour + (avg_pace.minute / 60))
```

```python
df.avg_pace.head(10)
```
