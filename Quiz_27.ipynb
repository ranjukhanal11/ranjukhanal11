```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
```

```python
plt.pie(x=df.groupby('workout_type').workout_type.agg('count'),labels = df.groupby('workout_type').workout_type.agg('count').index, wedgeprops={'edgecolor': 'black'}, counterclock=False, autopct='%.1f')
plt.show()
```

A mosaic plot is a subset of a stacked bar chart. The breadth of the column for two variables is proportional to the number of data in each level of the variable represented on the horizontal axis. The number of observations in the second variable inside each level of the first variable is equal to the vertically length of the bars.

In a mosaic, all of the columns of tiles have the same height, indicating 100%. Whereas sub-bars in a stacked bar graph indicate the joint frequency of a class of each of the 2 factors, each rectangles in a mosaic plot reflects the joint distribution of a class of each of the two variables.
