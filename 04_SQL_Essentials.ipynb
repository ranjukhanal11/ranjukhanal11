## Q3: Compute the average time for a trip based on starting point

```python
%%bigquery --project white-device-278509

select start_station_name ,avg(duration_minutes) as average from bigquery-public-data.austin_bikeshare.bikeshare_trips 
group by start_station_name
```

```python
%%bigquery --project white-device-278509

select start_station_name ,count(duration_minutes) as Total_Trips from bigquery-public-data.austin_bikeshare.bikeshare_trips 
group by start_station_name
```
